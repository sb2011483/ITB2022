##음주여부

음주여부

```python
sns.countplot(x="음주여부", data=df)
```

성별에 따른 음주여부

```python
sns.countplot(x="음주여부", data=df, hue="성별코드")
```

-> 여성보다 남성의 음주율이 더 높은 것을 알 수 있다.

연령대별 음주여부

```python
sns.boxplot(data=df, x="연령대코드(5세단위)", hue="음주여부")
```

-> 음주의 연령 범위, 중앙값이 12 즉, 55~59세인 걸 알 수 있다.

연령대별 음주여부에 따른 총콜레스테롤

```python
sns.lineplot(data=df, x="연령대코드(5세단위)", y="허리둘레", hue="음주여부")
```

-> 음주를 안 하는 사람보다 하는 사람의 허리둘레가 비교적 높은 것을 알 수 있다

```python
sns.lmplot(data=df, x="수축기혈압", y="이완기혈압", hue="음주여부")
```

-> 음주자의 수축,이완기 혈압이 비교적 더 높은 것을 알 수 있다
