##음주여부

**음주여부**

```python
sns.countplot(x="음주여부", data=df)
```

**성별에 따른 음주여부**

```python
sns.countplot(x="음주여부", data=df, hue="성별코드")
```

-> 여성보다 남성의 음주율이 더 높은 것을 알 수 있다.

**음주여부와 허리둘레**

```python
sns.lineplot(data=df, x="연령대코드(5세단위)", y="허리둘레", hue="음주여부")
```

-> 음주를 안 하는 사람보다 하는 사람의 허리둘레가 비교적 높은 것을 알 수 있다

**음주여부와 혈압**

```python
sns.lmplot(data=df, x="수축기혈압", y="이완기혈압", hue="음주여부")
```

-> 그래프 결과 음주와 혈압은 상관관계가 있다는 것을 알 수 있다.

**신장, 체중과 음주여부의 상관관계**

```python
sns.lmplot(data=df, x="신장(5Cm단위)", y="체중(5Kg 단위)", hue="음주여부")
```

```python
sns.lmplot(data=df, x="신장(5Cm단위)", y="체중(5Kg 단위)", hue="음주여부", col="성별코드")
```

-> 음주 여부에 따라 신장과 체중의 상관관계가 높은 것으로 보인다.
